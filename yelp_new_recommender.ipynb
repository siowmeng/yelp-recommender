```python
import numpy as np
import json
```

```python
dataDir = './data/yelp_dataset/'
dataFileNames = ['business', 'review']#, 'user', 'review']
```

```python
jsonDict = {}
for fname in dataFileNames:
    with open(dataDir + fname + '.json') as jsonFile:
        fContent = jsonFile.readlines()

    jsonDict[fname] = [json.loads(line) for line in fContent]
```

## Preprocess business

```python
jsonDict['business'][9]
```

```python
bizDict = {}
for jsonDoc in jsonDict['business']:
    bizDict[jsonDoc['business_id']] = {k: jsonDoc[k].strip() if type(jsonDoc[k]) is str else jsonDoc[k] \
                                               for k in ('name', 'city', 'state', 'stars', 'review_count')}
```

```python
bizCount = {}
for biz in jsonDict['business']:
    bizState = biz['state'].strip()
    bizCity = biz['city'].strip()
    if (bizState, bizCity) not in bizCount:
        bizCount[(bizState, bizCity)] = biz['review_count']
    else:
        bizCount[(bizState, bizCity)] += biz['review_count']
sorted(bizCount.items(), key = lambda kv: -kv[1])
```

## Preprocess review

```python
targetCities = [('NV', 'Las Vegas'), 
                ('AZ', 'Phoenix'), 
                ('NC', 'Charlotte'), 
                ('PA', 'Pittsburgh'), 
                ('WI', 'Madison')]
```

```python
reviewCity = {}

for i in range(len(jsonDict['review'])):
    
    cityTup = (bizDict[jsonDict['review'][i]['business_id']]['state'], 
               bizDict[jsonDict['review'][i]['business_id']]['city'])
    
    jsonDict['review'][i]['state'] = cityTup[0]
    jsonDict['review'][i]['city'] = cityTup[1]    
    
    if cityTup in targetCities:
        if cityTup in reviewCity:
            reviewCity[cityTup].append(jsonDict['review'][i])
        else:
            reviewCity[cityTup] = [jsonDict['review'][i]]
```
